# Feature encoding

### process of converting values to numerical values called feature encoding

#### '''categorically are generally divided into 3 types'''

##### 1) binary
###### yes or no

##### 2) Ordinal : specific ordered groups
###### low ,median,Mode
###### cold hot lava hot

##### 3) Nominal: Unordered groups
###### cat dpg tiger
###### pizza burger coke

```python
data={'col0':['mon','tue','wed','mon','fri','tue'],
      'col1':[1,2,3,2,5,1,],
      'col2':['True','False','True','False','True','False'],
      'col3':['red','blue','green','blue','yellow','green']}
print(data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)  ### convert to data frame or table form 

df
```

##### label encoder--- conerting binary and oridnaal to numeric data

```python
le=LabelEncoder()
df['col0']=le.fit_transform(df['col0'])
df
```

```python
import pandas as pd
data={'col0':['mon','tue','wed','mon','fri','tue'],
      'col1':[1,2,3,2,5,1,],
      'col2':['True','False','True','False','True','False'],
      'col3':['red','blue','green','blue','yellow','green']}
df1=pd.DataFrame(data)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in('col0','col2'):
    df1[i]=le.fit_transform(df1[i])
df1    
```

### One hot encoder
#### IS used to nominal to numeric data

```python
from sklearn.preprocessing import OneHotEncoder
One=OneHotEncoder(sparse_output=False)
df1['col3']=One.fit_transform(df1[['col3']])
df1
```

```python
from sklearn.preprocessing import OneHotEncoder 
enc=OneHotEncoder(sparse_output=False)
enc=enc.fit_transform(df1[['col3']]).astype('int')
encoded_colm=pd.DataFrame(enc)
df1=pd.concat([df1,encoded_colm],axis=1)
df1.drop('col3',axis=1,inplace=True)
df1
```

```python
from sklearn.preprocessing import OneHotEncoder 
enc=OneHotEncoder(sparse_output=False)
enc_col=enc.fit_transform(df1[['col3']]).astype('int')
encoded_colm=pd.DataFrame(enc_col,columns=enc.get_feature_names_out(['col3']))
df1=pd.concat([df1,encoded_colm],axis=1)
df1.drop('col3',axis=1,inplace=True)
df1
```

### FREQUENCY ENCODING: This method replace each category with its frequency or the count of how many times it appears in the data set

### '''value_counts(): This method counts the occurrences of each unique value in the 'nom_0' column.
#### The result is a Series where the index is the unique values from 'nom_0', 
#### and the values are the counts (frequencies) of those unique values.
#### This converts the Series to a dictionary. The keys of the dictionary are the unique values from 'nom_0', 
#### and the values are their correspondingÂ frequencies.'''

```python
fc = df1['col3'].value_counts().to_dict()
fc
```

```python
fc=df1['col3'].value_counts().to_dict()
df1['col3_encoded']=df1['col3'].map(fc)
df1
```
